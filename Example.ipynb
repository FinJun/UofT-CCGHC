```python
category_map = {
    'Line Impact': 'human', 'Unsafe Fire': 'human', 'Mechanical Failure': 'human',
    'Incendiary Device': 'human', 'Animals': 'nature', 'Winter Burning': 'nature',
    'Burning Substance': 'nature', 'Unclassified': 'nature', 'Friction Spark': 'nature',
    'Vehicle Fire': 'nature', 'Abandoned Fire': 'human', 'Arson Suspected': 'human',
    'Unattended Fire': 'human', 'Unpredictable Event': 'nature', 'Insufficient Buffer': 'human',
    'Permit Related': 'human', 'Hot Exhaust': 'human', 'Arson Known': 'human',
    'High Hazard': 'human', 'Insufficient Resources': 'human', 'Flammable Fluids': 'human'
}

df_4['true_cause'] = df_4['true_cause'].replace(category_map)
```

```python
df_4['true_cause'].fillna('nature', inplace=True)

print(df_4['true_cause'].unique())
```

```python
# Train, test data split

from pycaret.classification import *

model_setup = setup(data=df_4, target='true_cause', train_size=0.8, normalize=True, 
                    normalize_method='zscore', session_id=123)
```
